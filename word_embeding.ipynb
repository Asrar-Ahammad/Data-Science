```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# Sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
# Initialize vocabulary size.Increase in size increases processing time. 
voc_size = 10000
```

# One-Hot representation

```python
# To caputre index of 1 in one hot encoding. 
onehot_rep = [one_hot(words,voc_size) for words in sent]
```

```python
# Prints index of 1.    
# like : the - 3405, glass - 2916, of - 4033, milk - 9607
# Index value change everytime we run a program. 
print(onehot_rep)
```

# Word Embedding representaion

```python
# Inorder to make all the sentences to have same number of words we use padding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# Pre padding adding '0' at begining
sent_length = 8
embedded_docs = pad_sequences(onehot_rep, padding = 'pre', maxlen = sent_length)
print(embedded_docs)
```

```python
# Post padding adding '0' at ending
sent_length = 8
embedded_docs1 = pad_sequences(onehot_rep, padding = 'post', maxlen = sent_length)
print(embedded_docs1)
```

```python
# 10 fearure dimensions has been taken
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length = sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
# Every vocabulary is represented with 10 dimensions
model.predict(embedded_docs[0])
```
